# Local angle and crystallinity

Compute the local angle and crystallinity of a set of points. 
The nearest neighbours are identify using a threshold (KDTree in Scipy).
The local angle and cristallinity of an atom $i$ are computed from the local director field (see https://en.wikipedia.org/wiki/Hexatic_phase and SI of https://www.nature.com/articles/s41567-021-01429-3):
$$
    \psi_i= \frac{1}{N_i} \sum_j^{<i>} \exp(i N_\mathrm{ord} \theta_{ij}),
$$
where $\theta_{ij}$ is the angle of the bond vector $\mathbf{\delta}_{ij}=\mathbf{r}_j-\mathbf{r}_i$, $N_i$ is the number of nearest neighbour of $i$, and $N_\mathrm{ord}$ is the symmetry order of the lattice (6 for triangular, 4 for square).

The local angle is the computed as the argument of the complex function $\psi$.

The crystallinity as the absolute value of function $\psi$.

These snippets are implemented in the script ```local_angle-complex.py```

````python
def plt_cosmetic(ax):
    """Shortcut to adjust position plot"""
    ax.axhline(color='gray', ls=':', lw=1)
    ax.axvline(color='gray', ls=':', lw=1)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_aspect('equal')
````

````python
from tool_create_cluster import cluster_from_params, rotate
# --- Cluster parameters
params = {
    #'a1': [1,0], 'a2': [0,1], # square
    'a1': [1,0], 'a2': [-1/2, sqrt(3)/2], # triangular
    'cl_basis': [[0,0]], # No basis
    #'cl_basis': [[0,0], [-0.5, 0.28867513]], # Honeycomb
    'cluster_shape': 'circle', # cluster shape: circle, hexagonal, rectangle
    'N1': 20, 'N2': 20, # cluster size along lattice directions
    'theta': 15, 'pos_cm': [0,0] # rotation and shift
}
# Lattice spacing
R = np.linalg.norm(params['a1']) # assuming the two lattice vector are same length

# --- Create the cluster
pos = cluster_from_params(params)
N = pos.shape[0] # number of particles
print("Cluster %s of size N=%i" % (params['cluster_shape'], N))

# --- Honeycomb ONLY: align honeycomb to x axis, otherwise angle shifts of 30 deg
# pos = rotate(pos, 90)

# --- Rotate cluster
pos = rotate(pos, params['theta']) + params['pos_cm']
# --- Rotate central disk back to align, just to test
mask_disk = np.linalg.norm(pos, axis=1) < 5
pos[mask_disk] = rotate(pos[mask_disk], -params['theta']) 

# --- Add some noise to make it more interesting. Sayt 10% of lattice spacing
from numpy.random import uniform
noise = 0.1*R*(uniform(size=pos.shape)-1/2)*2 # between 0 and 1, scaled
pos += noise

# --- SHOW IT ---
plt.scatter(pos[~mask_disk,0], pos[~mask_disk,1], s=10)
plt.scatter(pos[mask_disk,0], pos[mask_disk,1], c='red', s=10)
plt_cosmetic(plt.gca())
plt.gcf().set_dpi(100)
plt.show()
````

````python
# Triangular lattice
d0 = R*np.sqrt(3)*0.8 # No basis: a bit less than second neighbour
# d0 = R*0.8 # # Honeycomb: a bit less than second neighbour
th_period = 60 # angular periodicity
Nnn0 = 6 # Symmetry order

# Square lattice
# d0 = R*np.sqrt(2)*0.8 # a bit less than second neighbour 
# th_period = 90
# Nnn0 = 4

# --- Identify nearest neighbour
from scipy.spatial import cKDTree
neigh_tree = cKDTree(pos, copy_data=True) # Neighour list object
nn = neigh_tree.query_pairs(d0, output_type='ndarray') # Nearest neighbours mask based on first config
pos_nn = pos[nn] # Position of nearest neighbours
````

````python
from matplotlib.colors import Normalize
from matplotlib.cm import Spectral
import matplotlib.cm as cmx
from scipy.spatial import cKDTree


# Complex orientational order function 
psi = np.zeros(pos.shape[0], dtype=np.complex128)

# -----------  Local angle ------------
norm = Normalize(-th_period/2, th_period/2) # bounds of the angle
cmap = cmx.ScalarMappable(norm=norm, cmap=Spectral) 

for cind in range(pos.shape[0]):
    mask_cnn = np.logical_or(nn[:,0] == cind, nn[:,1] == cind) # select the nearest neighbour of each atom
    for il in pos_nn[mask_cnn]:
        dd = il[1] - il[0] # bond vector
        c_angle = np.round(np.arctan2(dd[1], dd[0]), decimals=6) % (th_period*pi/180)
        psi[cind] += np.exp(1j*Nnn0*c_angle)
        # plot each bond vector one
        c_th = 180/pi*np.angle(np.exp(1j*Nnn0*c_angle))/Nnn0
        plt.plot([il[0][0], (il[0]+dd)[0]],[il[0][1], (il[0]+dd)[1]], color=cmap.to_rgba(c_th))
    psi[cind] /= pos_nn[mask_cnn].shape[0]
# local angle
avg_angle = 180/pi*np.angle(psi)/Nnn0

# show it
sc = plt.scatter(pos[:,0], pos[:,1], c=avg_angle, norm=norm, cmap='Spectral', s=20, zorder=10, ec='k', lw=0.3)
plt.colorbar(sc, label=r'$\theta=\arg(\psi)$')
plt_cosmetic(plt.gca())
plt.gcf().set_dpi(200)
plt.show()
# statistic of angles
print(np.average(avg_angle))
plt.hist(avg_angle)
plt.xlabel(r'$\theta$')
plt.ylabel(r'count')
plt.show()

# -----------  Crystallinity ------------
avg_crystal = np.abs(psi)
# plot bonds between particles
for cind in range(pos.shape[0]):
    mask_cnn = np.logical_or(nn[:,0] == cind, nn[:,1] == cind)
    for il in pos_nn[mask_cnn]:
        dd = il[1] - il[0]
        plt.plot([il[0][0], (il[0]+dd)[0]],[il[0][1], (il[0]+dd)[1]], 'k', lw=0.5)
# show it
crys_norm = Normalize(0,1) # cristallity is well normalised
sc = plt.scatter(pos[:,0], pos[:,1], c=avg_crystal, norm=crys_norm, cmap='magma', s=20, zorder=10, ec='k', lw=0.3)
plt.colorbar(sc, label=r'$|\psi|$')
plt_cosmetic(plt.gca())
plt.gcf().set_dpi(200)
plt.show()
# statistic of crystallinity
plt.hist(avg_crystal)
plt.xlabel(r'$|\psi|$')
plt.ylabel(r'count')
plt.show()
````
